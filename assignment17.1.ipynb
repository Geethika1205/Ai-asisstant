```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load the dataset assuming it's uploaded to the Colab environment
df_sales = pd.read_csv("/content/sales_data.csv")

print("Original Sales DataFrame head:")
display(df_sales.head())
print("\nOriginal Sales DataFrame info:")
df_sales.info()
```

```python
# Convert 'Sale_Date' column to datetime format
df_sales['Sale_Date'] = pd.to_datetime(df_sales['Sale_Date'])

print("DataFrame info after Date conversion:")
df_sales.info()
display(df_sales.head())
```

```python
# Create new features: month, year, day of week
df_sales['Month'] = df_sales['Sale_Date'].dt.month
df_sales['Year'] = df_sales['Sale_Date'].dt.year
df_sales['DayOfWeek'] = df_sales['Sale_Date'].dt.dayofweek # Monday=0, Sunday=6

print("DataFrame head with new time-based features:")
display(df_sales.head())
```

```python
# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Normalize the 'Sales_Amount' column
df_sales['Normalized_Sales'] = scaler.fit_transform(df_sales[['Sales_Amount']])

print("DataFrame head with Normalized Sales:")
display(df_sales.head())
```

```python
# Detect outliers using IQR method
Q1 = df_sales['Sales_Amount'].quantile(0.25)
Q3 = df_sales['Sales_Amount'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Lower bound for outliers: {lower_bound}")
print(f"Upper bound for outliers: {upper_bound}")

# Handle outliers by capping them
df_sales['Sales_Handled_Outliers'] = df_sales['Sales_Amount'].clip(lower=lower_bound, upper=upper_bound)

print("\nSales distribution before and after outlier handling:")
print("Original Sales min/max:", df_sales['Sales_Amount'].min(), df_sales['Sales_Amount'].max())
print("Outlier Handled Sales min/max:", df_sales['Sales_Handled_Outliers'].min(), df_sales['Sales_Handled_Outliers'].max())

display(df_sales.head())
```

```python
###TASK 4
# Step 1: Import libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Step 2: Load dataset
data = pd.read_csv("/content/Employee_Salary_Dataset.csv")

print("Original Data:")
print(data.head())
print("\nShape before preprocessing:", data.shape)


# ----------------------------------
# Step 3: Handle Missing Values
# ----------------------------------

# Fill numerical columns with median
numerical_cols = data.select_dtypes(include=['int64', 'float64']).columns

for col in numerical_cols:
    data[col].fillna(data[col].median(), inplace=True)

# Fill categorical columns with mode
categorical_cols = data.select_dtypes(include=['object']).columns

for col in categorical_cols:
    data[col].fillna(data[col].mode()[0], inplace=True)


# ----------------------------------
# Step 4: Standardize Department Names
# ----------------------------------

if 'Department' in data.columns:
    data['Department'] = data['Department'].str.lower().str.strip()


# ----------------------------------
# Step 5: Detect & Remove Salary Outliers (IQR)
# ----------------------------------

# Assume salary column name
salary_col = None

for col in data.columns:
    if 'salary' in col.lower():
        salary_col = col
        break

if salary_col:
    Q1 = data[salary_col].quantile(0.25)
    Q3 = data[salary_col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    print(f"\nSalary Outlier Bounds: {lower_bound} to {upper_bound}")

    # Remove outliers
    data = data[(data[salary_col] >= lower_bound) & (data[salary_col] <= upper_bound)]


# ----------------------------------
# Step 6: Label Encoding
# ----------------------------------

le = LabelEncoder()

# Encode Department
if 'Department' in data.columns:
    data['Department'] = le.fit_transform(data['Department'])

# Encode Job Location (try common names)
location_col = None
for col in data.columns:
    if 'location' in col.lower():
        location_col = col
        break

if location_col:
    data[location_col] = le.fit_transform(data[location_col].astype(str))


# ----------------------------------
# Step 7: Final Output
# ----------------------------------

print("\nProcessed Data:")
print(data.head())

print("\nShape after preprocessing:", data.shape)

# Save cleaned dataset
data.to_csv("processed_employee_salary.csv", index=False)

print("\n✅ Preprocessing Completed. File saved as 'processed_employee_salary.csv'")
```
