```python
# Python Class
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Salary: {self.salary}")

    def increase_salary(self, amount):
        self.salary += amount


# Run Python version
emp = Employee("John", 50000)
emp.display_info()

emp.increase_salary(5000)
print("\nAfter Salary Increase:")
emp.display_info()
```

```python
%%writefile Employee.java
class Employee {
    String name;
    double salary;

    // Constructor
    Employee(String name, double salary) {
        this.name = name;
        this.salary = salary;
    }

    void displayInfo() {
        System.out.println("Name: " + name);
        System.out.println("Salary: " + salary);
    }

    void increaseSalary(double amount) {
        salary += amount;
    }

    public static void main(String[] args) {
        Employee emp = new Employee("John", 50000);

        emp.displayInfo();

        emp.increaseSalary(5000);
        System.out.println("\nAfter Salary Increase:");
        emp.displayInfo();
    }
}
```

```python
# Python recursive factorial
def factorial(n):
    if n < 0:
        raise ValueError("Factorial not defined for negative numbers")
    if n in (0, 1):
        return 1
    return n * factorial(n - 1)

# Test
num = 5
print(f"Factorial of {num} is {factorial(num)}")

# More tests
for val in [0, 1, 3, 5, 7]:
    print(f"{val}! = {factorial(val)}")
```

```python
%%writefile factorial.cpp
#include <iostream>
using namespace std;

int factorial(int n) {
    if (n == 0 || n == 1)
        return 1;
    return n * factorial(n - 1);
}

int main() {
    int num = 5;
    cout << "Factorial of " << num << " is " << factorial(num) << endl;

    // Extra test cases
    int test[] = {0, 1, 3, 5, 7};
    for (int i = 0; i < 5; i++) {
        cout << test[i] << "! = " << factorial(test[i]) << endl;
    }

    return 0;
}
```

```python
import pandas as pd

# Create sample employee dataset
data = {
    'name': ['John', 'Alice', 'Bob', 'Emma', 'David'],
    'salary': [40000, 60000, 55000, 45000, 70000],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
}

df = pd.DataFrame(data)

print("Original Data:")
display(df)
```

```python
# Pandas equivalent
result = df.loc[df['salary'] > 50000, ['name', 'salary']]

print("Filtered Data:")
display(result)
```

```python
# Check if result is correct
expected_names = ['Alice', 'Bob', 'David']

if list(result['name']) == expected_names:
    print("✅ Correct Output")
else:
    print("❌ Something is wrong")
```

```python
# Python version (BeautifulSoup)

from bs4 import BeautifulSoup

# Sample HTML (simulate webpage)
html = """
<html>
  <body>
    <h1>Products</h1>
    <p class="price">100</p>
    <p class="price">200</p>
    <p class="price">300</p>
  </body>
</html>
"""

# Parse HTML
soup = BeautifulSoup(html, "html.parser")

# Extract prices
prices = [int(p.text) for p in soup.find_all("p", class_="price")]

print("Python Output:", prices)
```
